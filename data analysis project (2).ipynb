# India EV Market Data Analysis from 2001 - 2024

## EV Makers and their location of Manufacturing Plant

The Indian electric vehicle (EV) market has seen significant growth over the past decade, with various manufacturers setting up production units across the country. Below is the analysis of some of the most popular EV makers in India, along with their plant locations taken from the dataset and a brief description of their contributions to the EV industry .

```python
import pandas as pd

# Loading dataset
df = pd.read_csv("C:\\Users\\chand\\Downloads\\data analysis project\\EV__by_Place.csv")
print(df)
```

```python
import matplotlib.pyplot as plt

# Count of EV manufacturers per state
state_counts = df["State"].value_counts()   #count unique occurrences of value
print(state_counts)
```

```python
#potting the graph
plt.figure(figsize=(10,5))
state_counts.plot(kind="bar", color="green")
plt.xlabel("State")
plt.ylabel("Number of EV Manufacturers")
plt.title("State-wise Distribution of EV Manufacturers in India")
plt.xticks(rotation=45)
plt.show()
```

```python
# Count of EV manufacturers per city
city_counts = df["Place"].value_counts()
print(city_counts)
```

```python
plt.figure(figsize=(11,11))
city_counts.plot(kind="pie", autopct='%1.1f%%', colors=["cyan", "yellow", "green", "orange", "purple"])
plt.title("City-wise Distribution of EV Manufacturers")
plt.ylabel("")
plt.show()
```

## Total Operational Public Charging Station for EV available in each state.

## Total vehicles (includes electric and all other fuels) registered by different category from 2001 - Aug 2024

Below is the analysis of Total vehicles (includes electric and all other fuels) registered by different category from 2001 - Aug 2024


```python
# Loading dataset3

df3 = pd.read_csv("C:\\Users\\chand\\Downloads\\data analysis project\\Vehicle Class - All.csv")
print(df3)

print("\nSummary Statistics:")
print(df3.describe())
```

```python
# Bar Plot - Vehicle Class vs Total Registrations
plt.figure(figsize=(20,9))
plt.barh(df3['Vehicle Class'], df3['Total Registration'], color='teal')
plt.xlabel('Total Registrations')
plt.ylabel('Vehicle Class')
plt.title('Vehicle Class-wise Total Registrations (2001 - Aug 2024)')
plt.gca().invert_yaxis()  # Invert y-axis for better visualization
plt.show()
```

##### Cumulative Registration by Vehicle Class graph shows the total number of registered vehicles accumulated over time for each vehicle class.

```python
plt.figure(figsize=(10, 6))
plt.plot(df3['Vehicle Class'], df3['Total Registration'], marker='o', linestyle='-', color='darkblue')
plt.xticks(rotation=45)
plt.xlabel('Vehicle Class')
plt.ylabel('Total Registrations')
plt.title('Cumulative Vehicle Registrations by Category')
plt.grid(True)
plt.show()
```

It represents the sum of all registered vehicles belonging to a particular class (like Two-Wheelers, Three-Wheelers, Light Motor Vehicles, Heavy Goods Vehicles, etc.) from the start of data collection (2001) to the date available (Aug 2024).

The cumulative registration gives an idea of how many vehicles of each type have been manufactured and registered in India over the given period.

#### Conclusions:
The Two-Wheeler (NT) category remains the most dominant, reflecting the affordability and convenience of two-wheelers in Indian transportation.This data indicates the significant dependency on two-wheelers as a primary mode of transport in India.

Light Motor Vehicles (LMV) are also highly prevalent, indicating increased private vehicle ownership.

The rise of electric two-wheelers and small commercial vehicles could further influence these numbers.

The category "Other Than Mentioned Above" includes 11,26,398 registrations, likely representing special-purpose and less common vehicles.

In Future, Enhancing public transport and EV adoption while optimizing infrastructure for light and heavy goods vehicles to balance the transport ecosystem.

## Total electric vehicles manufactured by makers from 2015 - Aug 2024 with the vehicle class

```python
df5 = pd.read_csv("C:\\Users\\chand\\Downloads\\data analysis project\\ev_sales_by_makers_and_cat_15-24.csv")
```

```python
print("Data Overview:")
print(df5.info())
print("\n Missing Values:")
print(df5.isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
print(df5.describe())
```

```python
# Total Vehicles Manufactured
total_vehicles = df5['Total Registration'].sum()
print(f"\nTotal Electric Vehicles Manufactured (2015-2024): {total_vehicles}")
```

```python
# Top EV Manufacturers
top_makers = df5.groupby('Maker')['Total Registration'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 EV Manufacturers:")
print(top_makers)
```

```python
production_totals = {
    '2015': df5['2015'].sum(),
    '2016': df5['2016'].sum(),
    '2017': df5['2017'].sum(),
    '2018': df5['2018'].sum(),
    '2019': df5['2019'].sum(),
    '2020': df5['2020'].sum(),
    '2021': df5['2021'].sum(),
    '2022': df5['2022'].sum(),
    '2023': df5['2023'].sum(),
    '2024': df5['2024'].sum()
}
print(production_totals)
```

```python
# converting dictionary into series
yearly_growth = pd.Series(production_totals)
print(yearly_growth)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(yearly_growth.index, yearly_growth.values, marker='o', linestyle='-', color='blue', label='EV Manufacturing Growth')
plt.title('Year-wise Growth of Electric Vehicle Manufacturing (2015-2024)')
plt.xlabel('Year')
plt.ylabel('Number of Vehicles Manufactured')
plt.grid(True)
plt.legend()
plt.show()
```

### CONCLUSION

#### Line chart showing the total EV sales trend from 2015 to 2024.
The line chart shows slow EV growth from 2015-2019, followed by a sharp rise from 2020 onwards. The highest sales are in 2023-2024, indicating rapid market expansion and increased EV adoption. 

#### Steady Growth (2015-2019):
EV sales increased gradually but remained relatively low.

#### Sharp Surge (2020 Onward): 
A noticeable spike in sales, likely due to increased EV adoption, improved policies, and infrastructure.

#### Peak in Recent Years: 
2023 and 2024 (so far) show the highest sales, indicating continued market expansion.

```python
# Top 10 manufacturers by total EV sales
top10 = df5.groupby("Maker")["Total Registration"].sum().nlargest(10)

# Plot bar chart for top manufacturers
plt.figure(figsize=(12, 6))
top10.plot(kind="bar", color="green")

plt.xlabel("Manufacturer")
plt.ylabel("Total EV Sales")
plt.title("Top 10 EV Manufacturers (2015-2024)")
plt.xticks(rotation=75)
plt.grid(axis="y")
plt.show()
```

### Conclusion

#### Market Leaders:
The top 10 manufacturers contribute a significant share of total EV sales.
A few manufacturers dominate the industry, highlighting strong brand influence and production capacity.

#### High Sales Concentration:
The gap between the top manufacturers and others suggests a highly competitive market with a few dominant players.
Companies with early market entry or strong government support seem to have higher sales.

#### Emerging Players:
Some manufacturers may have recently entered the EV space but show rapid growth.
Future market shifts could bring new leaders as EV adoption increases.

```python
# Group sales by category for each year (excluding total column)
category_yearly_sales = df5.groupby("Cat").sum().iloc[:, :-1].T

# Plot stacked area chart for category-wise sales trend over time
plt.figure(figsize=(12, 6))
category_yearly_sales.plot(kind="area", stacked=True, colormap="viridis", alpha=0.8, figsize=(12, 6))

plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.title("EV Sales Trend by Category (2015-2024)")
plt.xticks(rotation=45)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
